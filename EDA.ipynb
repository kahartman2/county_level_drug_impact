```python
eda_map = eda_data
```

```python
eda_map.head()
```

```python
eda_map['state'] = eda_map['index'].apply(lambda x: x.split(','))
eda_map['state'] = eda_map['state'].apply(lambda x: x[1].replace(' ', ''))
```

```python
scl = [[0.0, 'rgb(242,240,247)'],[0.2, 'rgb(218,218,235)'],[0.4, 'rgb(188,189,220)'],\
            [0.6, 'rgb(158,154,200)'],[0.8, 'rgb(117,107,177)'],[1.0, 'rgb(84,39,143)']]

data = [ dict(
        type='choropleth',
        colorscale = scl,
        autocolorscale = False,
        locations = eda_map['state'],
        z = eda_map['prescriptions_per_100'],
        locationmode = 'USA-states',
#         text = df['text'],
        marker = dict(
            line = dict (
                color = 'rgb(255,255,255)',
                width = 2
            ) ),
        colorbar = dict(
            title = 'Per 100 People in 2016')
        ) ]

layout = dict(
        title = 'Prescription Opioids By State',
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showlakes = True,
            lakecolor = 'rgb(255, 255, 255)'),
             )
    
fig = dict( data=data, layout=layout )
iplot( fig, filename='d3-cloropleth-map' )
```
