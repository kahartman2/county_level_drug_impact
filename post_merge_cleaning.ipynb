###### As done in the initial data, functions to handle missing county-level data, so this method can be reused on other feature variables in which there is missing data. For every county in which there is missing data, fill in data based on the average value for the respective state.  

```python
#add values to lists corresponding to each state in missing_states_dict 
def find_state_values(state, column_name, missing_states_dict, dataframe):
    for i in range(len(dataframe)):
        if dataframe['index'][i][-2:] == state:
            if type(dataframe[column_name][i]) != str: 
                if dataframe[column_name][i] != 1000: 
                    missing_states_dict[state].append(dataframe[column_name][i])
    return missing_states_dict

#find average value of each list in missing_states dict and the resulting number is the average value for each state    
def find_avg(state, missing_states_dict):  
    missing_states_dict[state] = sum(missing_states_dict[state])/len(missing_states_dict[state])
    return missing_states_dict
    
#for any county with missing value, add the corresponding average value for respective state
def add_missing_val(state, column_name, missing_states_dict, dataframe):
    for i in range(len(dataframe)):
            if pd.isnull(dataframe[column_name].iloc[i]):
                dataframe.set_value(i, column_name, missing_states_dict[state])
            if dataframe[column_name][i]==1000:
                dataframe.set_value(i, column_name, missing_states_dict[state])
```

```python
#fill in null values with above functions for all numerical columns
for col in number_columns:
    null = merged_data.loc[merged_data[str(col)].isnull()]
    missing_counties = null['index'].value_counts().index.tolist()
    missing_states = [county[-2:] for county in missing_counties]

    missing_states_dict = {} 
    for state in missing_states: 
        missing_states_dict[state] = []
        missing_states_dict = find_state_values(state, str(col), missing_states_dict, post_float_conversion)

    for state in missing_states_dict: 
        find_avg(str(state), missing_states_dict)

    for state in missing_states:
        add_missing_val(state, str(col), missing_states_dict, post_float_conversion)
```
